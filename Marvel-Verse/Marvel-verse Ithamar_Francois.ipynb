```python
## Unedited main/starter file
OriginalMarvelDATA_DF = pd.DataFrame(marvel_chars)
#OriginalMarvelDATA_DF.head(5)


## Copying original list to create a main dataframe base DF 
MarvelUniverseDossier_DF = OriginalMarvelDATA_DF
print(MarvelUniverseDossier_DF.count())
MarvelUniverseDossier_DF.head(15)
```

```python
#------------------------------  CLEAN DATAFRAME  -----------------------------------------#

## Drop unneeded columns: df.drop(columns=['A', 'B'], inplace=True)
MarvelUniverseDossier_DF.drop(columns=['page_id','ID','GSM','APPEARANCES','Year'], inplace=True)  


## Rename columns
MarvelUniverseDossier_DF.rename(columns = 
{'name':'TARGET', 'urlslug':'DETAILED FILE', 'ALIGN':'ALIGNMENT', 'EYE':'EYECOLOR', 'HAIR':'HAIRCOLOR', 'ALIVE':'CURRENT STATUS', 'FIRST APPEARANCE':'FIRST OBSERVED'}, inplace=True)
## OR could do: MarvelUniverseDossier_DF = MarvelUniverseDossier_DF.rename(columns = 
##{'name':'TARGET', 'urlslug':'DETAILED FILE', 'ALIGN':'ALIGNMENT', 'EYE':'EYECOLOR', 'HAIR':'HAIRCOLOR', 'ALIVE':'CURRENT STATUS', 'FIRST APPEARANCE':'FIRST OBSERVED'})


## Remove 'Characters' from ALIGH/SEX/ALIVE columns
MarvelUniverseDossier_DF = MarvelUniverseDossier_DF.stack().str.replace("Characters",'').unstack()
MarvelUniverseDossier_DF = MarvelUniverseDossier_DF.fillna('REDACTED')


## Remove 'Eyes' from 'EYECOLOR' and 'Hair' from 'HAIRCOLOR'. Replace any 'No' (eg 'No Hair, No Eyes, Nothing etc) with None.
MarvelUniverseDossier_DF['EYECOLOR'] = MarvelUniverseDossier_DF['EYECOLOR'].str.replace("Eyes",'')
MarvelUniverseDossier_DF['EYECOLOR'] = MarvelUniverseDossier_DF['EYECOLOR'].str.replace("No",'None')

MarvelUniverseDossier_DF['HAIRCOLOR'] = MarvelUniverseDossier_DF['HAIRCOLOR'].str.replace("Hair",'')
MarvelUniverseDossier_DF['HAIRCOLOR'] = MarvelUniverseDossier_DF['HAIRCOLOR'].str.replace("No",'None')
MarvelUniverseDossier_DF['HAIRCOLOR'] = MarvelUniverseDossier_DF['HAIRCOLOR'].str.replace("Bald",'None')

## Merging the 'EYECOLOR' and 'HAIRCOLOR' data/columns into one new concatinated column called 'EYES/HAIR'. Dropping 'EYECOLOR' and 'HAIRCOLOR'
MarvelUniverseDossier_DF["EYES/HAIR"] = MarvelUniverseDossier_DF["EYECOLOR"] + "/ "+ MarvelUniverseDossier_DF["HAIRCOLOR"]
MarvelUniverseDossier_DF.drop(columns=['EYECOLOR','HAIRCOLOR'], inplace=True) 


## Remove single '\' instances from 'TARGET NAME' column
MarvelUniverseDossier_DF['TARGET'] = MarvelUniverseDossier_DF['TARGET'].str.replace("\\",'')

## Remove '(Earth-616)' and the () special characters from 'TARGET NAME' column values only (not WIKI URL's)
MarvelUniverseDossier_DF['TARGET'] = MarvelUniverseDossier_DF['TARGET'].str.replace(" \\(Earth-616\\)","")

## Remove all info contain within the ()'s from 'TARGET NAME' column only (ie '(Peter Parker)' in "Spider-Man (Peter Parker)")
#MarvelUniverseDossier_DF['TARGET'] = MarvelUniverseDossier_DF['TARGET'].str.replace(" \\([^()]{0,}\\)", "")
MarvelUniverseDossier_DF['TARGET'] = MarvelUniverseDossier_DF['TARGET'].str.replace("[\(\[].*?[\)\]]", "")

## Add "https://marvel.fandom.com/wiki/" to concatinate the values in "DETAILED FILE" to create a URL. 
MarvelUniverseDossier_DF['DETAILED FILE'] = MarvelUniverseDossier_DF['DETAILED FILE'].str.replace("\\\/",'https://marvel.fandom.com/wiki/')



## Drop potential missed duplicates and Null/NaN data fields of original csv data (before adding additional data)
MarvelUniverseDossier_DF = MarvelUniverseDossier_DF.dropna()
MarvelUniverseDossier_DF = MarvelUniverseDossier_DF.drop_duplicates()

## Create new copy of main dataframe for backup then continue updating main.
ArchiveMarvelDossier_DF = MarvelUniverseDossier_DF.copy()

## Original CSV dossier data is cleaned and prepped
print(MarvelUniverseDossier_DF.count())
MarvelUniverseDossier_DF.head(15)

```

```python
#------------------------------  CREATE & REORDER NEW COLUMNS  --------------------------------#

## Creating a list of new Columns that will be (hopefully) filled by API & Web-Scraping. Defaulting values to "CLASSIFIED" in case there are future 'Null/NaN' or no data retrieval.
RedactedColumns = ["CODENAME", "CASE ID", "REAL NAME", "KNOWN ALIASES", "MUGSHOT", "INTELLIGENCE", "STRENGTH", "SPEED", "DURABILITY", "ENERGY", "FIGHTING", "DOCUMENTED STORIES"]
MarvelUniverseDossier_DF[RedactedColumns] = "CLASSIFIED"
## OR can do: MarvelUniverseDossier_DF = ArchiveMarvelDossier_DF[RedactedColumns].copy()

## Reordering columns to the desired configuration now (before data is inserted) and passing it to the main dataframe
ReOrderedColumns = ["CASE ID", "TARGET", "CODENAME", "REAL NAME", "KNOWN ALIASES", "ALIGNMENT", "EYES/HAIR", "SEX", "CURRENT STATUS", "INTELLIGENCE", "STRENGTH", "SPEED", "DURABILITY", "ENERGY", "FIGHTING", "FIRST OBSERVED", "DOCUMENTED STORIES", "DETAILED FILE", "MUGSHOT"]

MarvelUniverseDossier_DF = MarvelUniverseDossier_DF.reindex(columns= ReOrderedColumns)

## Create copy of cleaned DF for future use
TestDossier_DF = MarvelUniverseDossier_DF.copy()

## Variable to be used in the API call results/output to keep track of the total Target count
TotalTargets = MarvelUniverseDossier_DF["TARGET"].count()


print(f"This Dossier contains preliminary information on {TotalTargets} 'entities of interest' and will now access 'CLASSIFIED' data...") 
print(f"Proceed with caution, Agent...   With great power, comes great responsibility...")

MarvelUniverseDossier_DF.head(5)
#TestDossier_DF.head(5)
```

```python
## Start of user facing "data scan"
print()
print(" ----------------------------- ")
print("  BEGINNING DOSSIER DOWNLOAD   ")
print(" ----------------------------- ")
print()

from bs4 import BeautifulSoup
import requests



## Index variable to iterate through generated citipy list
nextHero = 0

## Counter to keep track of every 50 succussful Target calls to increase setCounter by 1
recordCounter = 1

## Counter to keep track of how many sets of 50 successful Target API calls captured
setCounter = 0

## Counter to keep track of all times a data API attempt was made (successful or not)
allHeroCounter = 1

## Array of columns 'TARGET NAME' and 'DETAILED FILE'
HeroNameArray = MarvelUniverseDossier_DF["TARGET"].to_numpy()
HeroWikiURL = MarvelUniverseDossier_DF["DETAILED FILE"].to_numpy()


## List to hold values from columns 'NAME' and 'WIKI URL'
HeroNameList = []
HeroWikIList = []

hero_name = HeroNameArray[nextHero]
hero_wiki_url = HeroWikiURL[nextHero]

## Starting default CODENAME entry
CodeNameFromWiki = "Spider-Man"
APIFormatter = ""
HeroNameForAPI = ""

MarvelWiki = hero_wiki_url


# Retrieve page with the requests module
scrap_response = requests.get(MarvelWiki).text

# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(scrap_response, 'lxml')


## Initializing blank variables that will be used in loop
RealNameFromWiki = ""
AliasesFromWiki = ""
Biography = ""
Intelligence = ""
Strength = ""
Speed = ""
Durability = ""
Energy = ""
Fighting = ""



## For loop to go through however many random Hero were collected:
for nextHero in range(TotalTargets):
    ## Code will try to execute but be cognizant for Exception errors, as well as coded error API checks
   
   ## Exception safegaurd, in case API/WebScraping encounters an error, so loop can continue on to the next Target in dataframe
   try:
      
      ## for nextHero in range(len(HeroNameArray)):
      HeroNameList.append(HeroNameArray[nextHero])
      HeroWikIList.append(HeroWikiURL[nextHero])

      ## Variables created that keep track of current Target and their Wiki URL via index, in case
      hero_name = HeroNameArray[nextHero]
      hero_wiki_url = HeroWikiURL[nextHero]
      

      ## Boolean value to indicate when API loop/calls should stop after third 'HeroNameForAPI' attempt
      Response = True

         
      ## URL of page to be scraped
      MarvelWiki = hero_wiki_url

      ## Retrieve page with the requests module
      scrap_response = requests.get(MarvelWiki).text
      ## Create BeautifulSoup object; parse with 'html.parser'
      soup = BeautifulSoup(scrap_response, 'lxml')


      ## This webscrapes the Wiki page/current url and gets Target information from the right hand "table"/infobox html object
      infobox = soup.find('div', class_="infobox")
      CodeNameFromWiki = infobox.a.text
      RealNameFromWiki = infobox.find('div', style="width:60%;text-align:left;float:left").text.strip()
      AliasesFromWiki = infobox.find('td', style="margin:0.5em;font-size:12px").text.strip()
      
      
      ## Update dataframe columns with the corresponding values that is webscraped
      MarvelUniverseDossier_DF.loc[nextHero, "CODENAME"] = CodeNameFromWiki
      MarvelUniverseDossier_DF.loc[nextHero, "REAL NAME"] = RealNameFromWiki
      MarvelUniverseDossier_DF["REAL NAME"] = MarvelUniverseDossier_DF["REAL NAME"].str.replace("[\(\[].*?[\)\]]", "")  ## 'Cleans the returned names/aliases of HTML footnote tags to just leave text.

      MarvelUniverseDossier_DF.loc[nextHero, "KNOWN ALIASES"] = AliasesFromWiki
      MarvelUniverseDossier_DF["KNOWN ALIASES"] = MarvelUniverseDossier_DF["KNOWN ALIASES"].str.replace("[\(\[].*?[\)\]]", "")
      
           
      
      ## Webscrape of the Power Grid "table" that is located in the Wiki page under 'Powers and Abilities' using html table read function.
      ## Location of Power grid table in current Wiki page
      PowerGrid = f"{soup.find('table', align='right')}"
      PowerTable = pd.read_html(PowerGrid)
      Powers = PowerTable[0]

      
      ## Needed to drop the first column (that contains strings) in order to eventually convert the new df into "float64" values for consistency
      Powers.drop(columns=[0], inplace=True)
      Powers = Powers[0:6].astype(float)
      
      ## Each df columns integer values are added to then attach to respective variables
      Stats = Powers.sum(axis=1)
      Intelligence = Stats[0]
      Strength = Stats[1]
      Speed = Stats[2]
      Durability = Stats[3]
      Energy = Stats[4]
      Fighting = Stats[5]

      ## Update main dataframe with the values extracted from Power Grid table
      MarvelUniverseDossier_DF.loc[nextHero, "INTELLIGENCE"] = Intelligence
      MarvelUniverseDossier_DF.loc[nextHero, "STRENGTH"] = Strength
      MarvelUniverseDossier_DF.loc[nextHero, "SPEED"] = Speed
      MarvelUniverseDossier_DF.loc[nextHero, "DURABILITY"] = Durability
      MarvelUniverseDossier_DF.loc[nextHero, "ENERGY"] = Energy
      MarvelUniverseDossier_DF.loc[nextHero, "FIGHTING"] = Fighting
      
      ## End of current URL webscrape


      ## API function

      ## Stores original 'Codename' values as we may have to edit it within loop
      OrigWikiCodeName = CodeNameFromWiki
      
      ## Take Codename and make the string consistent and all UPPERCASE after saving it (as a title case to a new variable that will display in API return results)
      CodeNameFromWiki.title()
      ReturnedWikiName = CodeNameFromWiki

      CodeNameFromWiki.upper()
      

      ## An issue that is arising is a conflict in how the Marvel.com API specifically needs a character's name submitted. 
      ## At times the same name will be denied depending on if it contains a space, a "-", or is one word (ie IceMan vs. Ice-Man vs Ice Man)
      ## In order to try to get around that, I'm attempting to submit multiple versions of the Codename to the API call in case there is call that wont work.
      

      ## Making a copy of CodeName to be edited/manipulated
      APIFormatter = CodeNameFromWiki

      ## Using new variable for URL calls, replaces a " " (in a Codename) with what the API URL will need for a name (ie 'Ice Man' => 'Ice%20Man')
      HeroNameForAPI = APIFormatter.replace(" ", "%20")


      
      ## Creating API URL dynamically for every Hero json call/response in HeroNameArray/'CodeName' list
      baseURL = "https://gateway.marvel.com:443/v1/public/characters?name="

      queryURL = (f"{baseURL}{HeroNameForAPI}{Marvel_API}")
      
      JsonHEROResponse = requests.get(queryURL).json()


      ## Code will try to execute, but be cognizant for Exception errors and various CodeName API checks
      try:
         
         ## API Call Limiter
         ## recordCounter will reset itself, and add 1 to the "Set" display, after every 50 successful entries. 
           
         if recordCounter == 51:
            recordCounter = 0
            setCounter += 1

         ## Designed to break out of the API call/loop after 150 successful updates (SetCounter = 3), but that variable can be commented out/edited to allow more Targets to be called.
         if setCounter == 2:  ## Scans will take 5mins at SetCounter == 3 approx
            print(f'Download aborted! CEREBRO lockout system triggered, X-Men trace initiated. Leave this area!! Commence connection log purge...')
            print(f'... Last recovered intel for {hero_name}, can be reviewed here: "{hero_wiki_url}" ') 
            break       ## Setting a manual break at 200 API calls (4sets of 50calls) to limit infinite loops of large scans. Can be removed if needed.
               
               
               
               ## API CALL/RESPONSE CHECK
               ## IF API response returns an incomplete dataset ('count = 0' OR 'total = 0')  
               ## this Code will return an error message (using originally returned Codename, and reattempt with a tweeked Codename
         if JsonHEROResponse["data"]["count"] == '0' or \
            JsonHEROResponse["data"]["total"] == '0': #or \
            #JsonHEROResponse["code"] != "200":

            Response = True
            print()
            print(f'Access Denied! Intel on "{ReturnedWikiName}" is not recognized by system. Trying again with different name, please wait...')
            print()

            
            ## if Respose remains True, API check will try again with new Codename attempt.
            if Response == True:
               
               ## Example) ReturnedWikiName = Giant Man  //  HeroNameForAPI = GIANT%20MAN // APIFormatter = GIANT MAN
               ## Attempt to be bame by changin values to now be ReturnedWikiName = Giant-Man  //  HeroNameForAPI = GIANT-MAN // APIFormatter = GIANT-MAN 
               APIFormatter = ReturnedWikiName.title()
               HeroNameForAPI = APIFormatter.replace(" ", "-")
               ReturnedWikiName = HeroNameForAPI
               HeroNameForAPI.upper()

               ## ReAttmept call with new ReturnWiki value (if it applies)
               queryURL = (f"{baseURL}{HeroNameForAPI}{Marvel_API}")
               JsonHEROResponse = requests.get(queryURL).json()

               if JsonHEROResponse["data"]["count"] == '0' or \
                  JsonHEROResponse["data"]["total"] == '0': #or \
                  #JsonHEROResponse["code"] != "200":
                           
                  
                  print(f'Access Denied! Intel on "{ReturnedWikiName}" is not recognized by system. Trying again with different name, please wait...')
                  print()               
                  
                  ## Same as above, just re-attempting the call with a different Target name
                  if Response == True:
                     
                     ## Example) ReturnedWikiName = Spider-Man  //  HeroNameForAPI = SPIDERMAN // APIFormatter = SPIDERMAN
                     APIFormatter = ReturnedWikiName.title()
                     HeroNameForAPI = APIFormatter.replace("-", "")
                     ReturnedWikiName = HeroNameForAPI
                     HeroNameForAPI.upper()

                     ## Response set False which will be the last API attempt condition with new variable
                     Response = False
                     queryURL = (f"{baseURL}{HeroNameForAPI}{Marvel_API}")
                     JsonHEROResponse = requests.get(queryURL).json()
                  
                     if JsonHEROResponse["data"]["count"] == '0' or \
                        JsonHEROResponse["data"]["total"] == '0': #or \
                        #JsonHEROResponse["code"] != "200":
                        
                        ## Final error message notifying user that we will move on to a new Target due to API fail
                        print(f'Access Denied! 3rd Attempt for "{ReturnedWikiName}" yielded lockout warning. Incomplete dossier update, aborting Target entry...')
                        print(f'Available intel for {hero_name}, that was recovered, can be reviewed here: "{hero_wiki_url}" ')
                  
                  
                     ## A successful API call, return Response to False (to stop API check/loop, update the dataframe with values extracted from Marvel API. Update the CodeName with whichever version worked.   
                     else:
                        Response = False
                        MarvelUniverseDossier_DF.loc[nextHero, "CODENAME"] = CodeNameFromWiki
                        MarvelUniverseDossier_DF.loc[nextHero, "CASE ID"] = JsonHEROResponse["data"]["results"][0]['id']
                        MarvelUniverseDossier_DF.loc[nextHero, "DOCUMENTED STORIES"] = JsonHEROResponse["data"]["results"][0]['urls'][2]['url']
                        
                        ## API Photo is broken into two sections: url + extension (ie: http://xxx  &  .jpg)
                        ## Need to combine both values and concatinate a full url address using variable to hold each part
                        ResponseScavenger1 = JsonHEROResponse["data"]["results"][0]['thumbnail']['path']
                        ResponseScavenger2 = JsonHEROResponse["data"]["results"][0]['thumbnail']['extension']
                        ResponseScavengerX = f"{ResponseScavenger1}.{ResponseScavenger2}"
                        ## Update MUGSHOT column with url                      
                        MarvelUniverseDossier_DF.loc[nextHero, "MUGSHOT"] = ResponseScavengerX
                        
                        ## Print succesful call message and go to the next Target. Display progress of updates via counters and current Hero and wiki url in the print results.
                        print(f" |  Downloaded intel summary on {hero_name}... ({ recordCounter } of Set { setCounter })...  Target {allHeroCounter} of {TotalTargets} total Targets updated.")
                        print(f" |  Detailed Target information for {hero_name} can be reviewed here: '{hero_wiki_url}' ")
                        print()

                  else:
                     ## Iterate counts
                     nextHero += 1
                     recordCounter += 1
                     allHeroCounter += 1
                     pass
               
               
               ## Same as above, updates the dataframe values after successful API call and then iterates to next Target
               else:
                  MarvelUniverseDossier_DF.loc[nextHero, "CODENAME"] = CodeNameFromWiki
                  MarvelUniverseDossier_DF.loc[nextHero, "CASE ID"] = JsonHEROResponse["data"]["results"][0]['id']
                  MarvelUniverseDossier_DF.loc[nextHero, "DOCUMENTED STORIES"] = JsonHEROResponse["data"]["results"][0]['urls'][2]['url']
                  ResponseScavenger1 = JsonHEROResponse["data"]["results"][0]['thumbnail']['path']
                  ResponseScavenger2 = JsonHEROResponse["data"]["results"][0]['thumbnail']['extension']
                  ResponseScavengerX = f"{ResponseScavenger1}.{ResponseScavenger2}"
                  MarvelUniverseDossier_DF.loc[nextHero, "MUGSHOT"] = ResponseScavengerX
                  print(f" |  Downloaded intel summary on {hero_name}... ({ recordCounter } of Set { setCounter })...  Target {allHeroCounter} of {TotalTargets} total Targets updated.")
                  print(f" |  Detailed Target information for {hero_name} can be reviewed here: '{hero_wiki_url}' ")
                  print()

                  


            else:
               ## Iterate counts
               nextHero += 1
               recordCounter += 1
               allHeroCounter += 1
               pass   



           

         else:
            # Update DF with new character data from the Marvel API (ie Marvel/API ID# (CASE ID), Comics featuring character (DOCUMENTED STORIES), Picture .jpg url (MUGSHOT))
            MarvelUniverseDossier_DF.loc[nextHero, "CODENAME"] = CodeNameFromWiki
            MarvelUniverseDossier_DF.loc[nextHero, "CASE ID"] = JsonHEROResponse["data"]["results"][0]['id']
            MarvelUniverseDossier_DF.loc[nextHero, "DOCUMENTED STORIES"] = JsonHEROResponse["data"]["results"][0]['urls'][2]['url']
            ResponseScavenger1 = JsonHEROResponse["data"]["results"][0]['thumbnail']['path']
            ResponseScavenger2 = JsonHEROResponse["data"]["results"][0]['thumbnail']['extension']
            ResponseScavengerX = f"{ResponseScavenger1}.{ResponseScavenger2}"
            MarvelUniverseDossier_DF.loc[nextHero, "MUGSHOT"] = ResponseScavengerX
            print(f" |  Downloaded intel summary on {hero_name}... ({ recordCounter } of Set { setCounter })...  Target {allHeroCounter} of {TotalTargets} total Targets updated.")
            print(f" |  Detailed Target information for {hero_name} can be reviewed here: '{hero_wiki_url}' ")
            print()
            ## Iterate counts
            nextHero += 1
            recordCounter += 1
            allHeroCounter += 1

      
      ## In the event of an indexerror/incomplet API, no data will be marked and loop/count will be iterated   
      except IndexError:
         print()
         print(f'Summary update for {hero_name} denied... CEREBRO is jammed signal, location trace begun... Aborting entry and connection kernel...')
         print()
         allHeroCounter += 1
         nextHero += 1
         pass
   
   
   ## In the event of a bad URL for the Marvel WIKI, continue to the next Target
   except:
      print(f'Corrupted intel on {hero_name}... Aborting entry and moving on...')
      pass


## End of "dossier update"
print()
print(" ----------------------------- ")
print("   DOSSIER UPDATE COMPLETE     ")
print(" ----------------------------- ")
print()

```
